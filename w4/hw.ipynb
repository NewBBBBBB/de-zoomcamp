```python
SELECT
    pickup_zone,
    SUM(revenue_monthly_total_amount) AS total_revenue
FROM prod.monthly_revenue_per_location
WHERE service_type = 'Green'
  AND revenue_month >= DATE '2020-01-01'
  AND revenue_month <  DATE '2021-01-01'
GROUP BY pickup_zone
ORDER BY total_revenue DESC
LIMIT 1;
```

```python
SELECT
      SUM(total_monthly_trips) AS total_revenue
  FROM prod.monthly_revenue_per_location
  WHERE service_type = 'Green'
    AND revenue_month >= DATE '2019-10-01'
    AND revenue_month <  DATE '2019-11-01'
  ORDER BY total_revenue DESC
  LIMIT 1;
```
